Para implementar um sistema de gerenciamento de biblioteca em Python que atenda aos requisitos fornecidos, podemos utilizar uma estrutura básica orientada a objetos. Vamos definir classes para representar os usuários, os livros e o próprio sistema de gerenciamento. Abaixo está um esboço do código que implementa essas funcionalidades:

### 1. Classes Principais

- **`Livro`**: Representa um livro no acervo.
- **`Usuario`**: Representa um usuário genérico.
- **`Administrador`**: Herda de `Usuario` e possui permissões especiais.
- **`Cliente`**: Herda de `Usuario`.
- **`SistemaBiblioteca`**: Gerencia o acervo de livros e as operações de empréstimo e devolução.

### Código do Sistema de Gerenciamento de Biblioteca

```python
from datetime import datetime, timedelta

class Livro:
    def __init__(self, titulo, autor, isbn):
        self.titulo = titulo
        self.autor = autor
        self.isbn = isbn
        self.disponivel = True
        self.data_emprestimo = None
        self.data_devolucao_prevista = None
        self.emprestado_para = None

    def emprestar(self, usuario, data_emprestimo):
        if self.disponivel:
            self.disponivel = False
            self.data_emprestimo = data_emprestimo
            self.data_devolucao_prevista = data_emprestimo + timedelta(days=5)
            self.emprestado_para = usuario
            print(f"Livro '{self.titulo}' emprestado para {usuario.nome}.")
        else:
            print(f"Livro '{self.titulo}' não está disponível para empréstimo.")

    def devolver(self, data_devolucao):
        if not self.disponivel:
            if data_devolucao > self.data_devolucao_prevista:
                print(f"Livro '{self.titulo}' devolvido com atraso!")
            self.disponivel = True
            self.data_emprestimo = None
            self.data_devolucao_prevista = None
            self.emprestado_para = None
            print(f"Livro '{self.titulo}' devolvido com sucesso.")
        else:
            print(f"Livro '{self.titulo}' não está emprestado.")

class Usuario:
    def __init__(self, nome, tipo):
        self.nome = nome
        self.tipo = tipo

class Administrador(Usuario):
    def __init__(self, nome):
        super().__init__(nome, tipo='Administrador')

class Cliente(Usuario):
    def __init__(self, nome):
        super().__init__(nome, tipo='Cliente')

class SistemaBiblioteca:
    def __init__(self):
        self.acervo = []

    def adicionar_livro(self, titulo, autor, isbn):
        novo_livro = Livro(titulo, autor, isbn)
        self.acervo.append(novo_livro)
        print(f"Livro '{titulo}' adicionado ao acervo.")

    def consultar_acervo(self):
        print("Acervo da Biblioteca:")
        for livro in self.acervo:
            status = "Disponível" if livro.disponivel else "Emprestado"
            print(f"Titulo: {livro.titulo}, Autor: {livro.autor}, ISBN: {livro.isbn}, Status: {status}")

    def realizar_emprestimo(self, admin, isbn, usuario):
        if not isinstance(admin, Administrador):
            print("Apenas um administrador pode realizar um empréstimo.")
            return
        for livro in self.acervo:
            if livro.isbn == isbn:
                livro.emprestar(usuario, datetime.now())
                return
        print(f"Livro com ISBN {isbn} não encontrado no acervo.")

    def realizar_devolucao(self, admin, isbn):
        if not isinstance(admin, Administrador):
            print("Apenas um administrador pode realizar uma devolução.")
            return
        for livro in self.acervo:
            if livro.isbn == isbn:
                livro.devolver(datetime.now())
                return
        print(f"Livro com ISBN {isbn} não encontrado no acervo.")

# Exemplo de uso
sistema = SistemaBiblioteca()
admin = Administrador("Admin 1")
cliente = Cliente("Cliente 1")

# Adicionando livros ao acervo
sistema.adicionar_livro("Python para Iniciantes", "Autor A", "123456789")

# Consultando acervo

# Consultando acervo após empréstimo

# Realizando devolução
sistema.realizar_devolucao(admin, "123456789")

# Consultando acervo após devolução
sistema.consultar_acervo()
```

### Explicação:

- **Livros**: A classe `Livro` mantém informações sobre cada livro, incluindo se está disponível para empréstimo e para quem está emprestado.
  
- **Usuários**: Existem dois tipos de usuários: `Administrador` (que pode realizar empréstimos e devoluções) e `Cliente` (que pode apenas consultar o acervo).

- **SistemaBiblioteca**: Esta classe gerencia a coleção de livros e as operações de empréstimo e devolução, garantindo que apenas administradores possam executar essas operações.

### Como Usar:

1. **Adicionar Livros**: Use `adicionar_livro` para adicionar novos livros ao acervo.
2. **Consultar Acervo**: Use `consultar_acervo` para listar todos os livros com seus status.
3. **Realizar Empréstimo/Devolução**: Use `realizar_emprestimo` e `realizar_devolucao`, respectivamente, para gerenciar os empréstimos.

````python
from datetime import datetime, timedelta

class Livro:
    def __init__(self, titulo, autor, isbn):
        self.titulo = titulo
        self.autor = autor
        self.isbn = isbn
        self.disponivel = True
        self.data_emprestimo = None
        self.data_devolucao_prevista = None
        self.emprestado_para = None

    def emprestar(self, usuario, data_emprestimo):
        if self.disponivel:
            self.disponivel = False
            self.data_emprestimo = data_emprestimo
            self.data_devolucao_prevista = data_emprestimo + timedelta(days=5)
            self.emprestado_para = usuario
            print(f"Livro '{self.titulo}' emprestado para {usuario.nome}.")
        else:
            print(f"Livro '{self.titulo}' não está disponível para empréstimo.")

    def devolver(self, data_devolucao):
        if not self.disponivel:
            if data_devolucao > self.data_devolucao_prevista:
                print(f"Livro '{self.titulo}' devolvido com atraso!")
            self.disponivel = True
            self.data_emprestimo = None
            self.data_devolucao_prevista = None
            self.emprestado_para = None
            print(f"Livro '{self.titulo}' devolvido com sucesso.")
        else:
            print(f"Livro '{self.titulo}' não está emprestado.")

class Usuario:
    def __init__(self, nome, tipo):
        self.nome = nome
        self.tipo = tipo

class Administrador(Usuario):
    def __init__(self, nome):
        super().__init__(nome, tipo='Administrador')

class Cliente(Usuario):
    def __init__(self, nome):
        super().__init__(nome, tipo='Cliente')

class SistemaBiblioteca:
    def __init__(self):
        self.acervo = []

    def adicionar_livro(self, titulo, autor, isbn):
        novo_livro = Livro(titulo, autor, isbn)
        self.acervo.append(novo_livro)
        print(f"Livro '{titulo}' adicionado ao acervo.")

    def consultar_acervo(self):
        print("Acervo da Biblioteca:")
        for livro in self.acervo:
            status = "Disponível" if livro.disponivel else "Emprestado"
            print(f"Titulo: {livro.titulo}, Autor: {livro.autor}, ISBN: {livro.isbn}, Status: {status}")

    def realizar_emprestimo(self, admin, isbn, usuario):
        if not isinstance(admin, Administrador):
            print("Apenas um administrador pode realizar um empréstimo.")
            return
        for livro in self.acervo:
            if livro.isbn == isbn:
                livro.emprestar(usuario, datetime.now())
                return
        print(f"Livro com ISBN {isbn} não encontrado no acervo.")

    def realizar_devolucao(self, admin, isbn):
        if not isinstance(admin, Administrador):
            print("Apenas um administrador pode realizar uma devolução.")
            return
        for livro in self.acervo:
            if livro.isbn == isbn:
                livro.devolver(datetime.now())
                return
        print(f"Livro com ISBN {isbn} não encontrado no acervo.")

# Exemplo de uso
sistema = SistemaBiblioteca()
admin = Administrador("Admin 1")
cliente = Cliente("Cliente 1")

# Adicionando livros ao acervo
from datetime import datetime, timedelta

class Livro:
    def __init__(self, titulo, autor, isbn):
        self.titulo = titulo
        self.autor = autor
        self.isbn = isbn
        self.disponivel = True
        self.data_emprestimo = None
        self.data_devolucao_prevista = None
        self.emprestado_para = None

    def emprestar(self, usuario, data_emprestimo):
        if self.disponivel:
            self.disponivel = False
            self.data_emprestimo = data_emprestimo
            self.data_devolucao_prevista = data_emprestimo + timedelta(days=5)
            self.emprestado_para = usuario
            print(f"Livro '{self.titulo}' emprestado para {usuario.nome}.")
        else:
            print(f"Livro '{self.titulo}' não está disponível para empréstimo.")

    def devolver(self, data_devolucao):
        if not self.disponivel:
            if data_devolucao > self.data_devolucao_prevista:
                print(f"Livro '{self.titulo}' devolvido com atraso!")
            self.disponivel = True
            self.data_emprestimo = None
            self.data_devolucao_prevista = None
            self.emprestado_para = None
            print(f"Livro '{self.titulo}' devolvido com sucesso.")
        else:
            print(f"Livro '{self.titulo}' não está emprestado.")

class Usuario:
    def __init__(self, nome, tipo):
        self.nome = nome
        self.tipo = tipo

class Administrador(Usuario):
    def __init__(self, nome):
        super().__init__(nome, tipo='Administrador')

class Cliente(Usuario):
    def __init__(self, nome):
        super().__init__(nome, tipo='Cliente')

class SistemaBiblioteca:
    def __init__(self):
        self.acervo = []

    def adicionar_livro(self, titulo, autor, isbn):
        novo_livro = Livro(titulo, autor, isbn)
        self.acervo.append(novo_livro)
        print(f"Livro '{titulo}' adicionado ao acervo.")

    def consultar_acervo(self):
        print("Acervo da Biblioteca:")
        for livro in self.acervo:
            status = "Disponível" if livro.disponivel else "Emprestado"
            print(f"Titulo: {livro.titulo}, Autor: {livro.autor}, ISBN: {livro.isbn}, Status: {status}")

    def realizar_emprestimo(self, admin, isbn, usuario):
        if not isinstance(admin, Administrador):
            print("Apenas um administrador pode realizar um empréstimo.")
            return
        for livro in self.acervo:
            if livro.isbn == isbn:
                livro.emprestar(usuario, datetime.now())
                return
        print(f"Livro com ISBN {isbn} não encontrado no acervo.")

    def realizar_devolucao(self, admin, isbn):
        if not isinstance(admin, Administrador):
            print("Apenas um administrador pode realizar uma devolução.")
            return
        for livro in self.acervo:
            if livro.isbn == isbn:
                livro.devolver(datetime.now())
                return
        print(f"Livro com ISBN {isbn} não encontrado no acervo.")

# Exemplo de uso
sistema = SistemaBiblioteca()
admin = Administrador("Admin 1")
cliente = Cliente("Cliente 1")

# Adicionando livros ao acervo
sistema.adicionar_livro("Python para Iniciantes", "Autor A", "123456789")
sistema.adicionar_livro("Aprenda Java", "Autor B", "987654321")

# Consultando acervo
sistema.consultar_acervo()

# Realizando empréstimo
sistema.realizar_emprestimo(admin, "123456789", cliente)

# Consultando acervo após empréstimo
sistema.consultar_acervo()

# Realizando devolução
sistema.realizar_devolucao(admin, "123456789")

# Consultando acervo após devolução
sistema.consultar_acervo()


# Consultando acervo
sistema.consultar_acervo()

# Realizando empréstimo
sistema.realizar_emprestimo(admin, "123456789", cliente)

# Consultando acervo após empréstimo
sistema.consultar_acervo()

# Realizando devolução
sistema.realizar_devolucao(admin, "123456789")

#Consultando o acervo após devolução
sistema.consultar_acervo()
````
